```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
xab = []
for i in range(8):
    xab.append(i*15.14)
xbc = [105.98,132.52]
xcd = [132.52,156.42,180.32,204.22,228.12,252.02,257.92,293.84,300]

yab = [0,5.47,7.72,8.98,9.53,10.02,9.75,15.06]
ybc = [15.06,36.17]
ycd = [36.17,37.39,36.24,33.99,25.33,11.83,5.91,1.04,6.52]
plt.plot(xab,yab,'r',xbc,ybc,'g',xcd,ycd,'b')
```

# Parte A

```python
## Parte AB
xi = np.arange(0,xab[-1],0.1)
n = interpolate.splrep(xab,yab)
s = lambda xn: interpolate.splev(xn,n)
si =  []
for i in xi:
    si.append(s(i))

## Parte CD
xg = np.arange(xcd[0],xcd[-1],0.1)
g = interpolate.splrep(xcd,ycd)
s = lambda xn: interpolate.splev(xn,g)
sg =  []
for i in xg:
    sg.append(s(i))

plt.plot(xab,yab,'ko',xi,si,'b',xbc,ybc,'g',xg,sg,'r',xcd,ycd,'ko')
```
