```python
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x = [2,2.5,4,4.5,6]
y = [1.623,1.855,2.041,2.333,2.561]
plt.plot(x,y,'*')
```

```python
t = interpolate.splrep(x,y)
k = lambda x: interpolate.splev(x,t)
```

```python
xl = x[4]-x[0]
intervalos = 25
yl = y[4]-y[0]
dx = xl / intervalos
dy = yl / intervalos
xnew = []
ynew = []

for p in range(intervalos):
    xn = x[0] + p*dx
    ynew.append(y[0]+p*dy)
    xnew.append(k(xn))

plt.plot(xnew,ynew)
print (xnew,ynew)
```
